```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from time import sleep
```

```python
path ="./Lidar_Data/2021/"

filelist = []
for root, dirs, files in os.walk(path):
    for file in files:
        if(file.endswith(".csv")):
            filelist.append(os.path.join(root,file))
```

```python
len(filelist)
```

```python
plt.figure(figsize=(12,6))
for filename in filelist:
    df = pd.read_csv(filename)
    plt.plot(df['time'],df['photon_counts'],label=filename)
plt.xlabel("Time")
plt.ylabel("Photon Counts")
plt.title("MCS Data 2021, 575 files")
plt.show()
#plt.legend()
```
